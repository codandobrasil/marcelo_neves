```python
import random

# Lista de palavras que irão ser sorteadas no jogo
palavras = ['amor', 'abacaxi', 'jamanta', 'comida', 'homem', 'mulher', 'computador','variavel']

#Escolha das palavras pelo método randon
palavra_escolhida = random.choice(palavras)

#Criação de strings para representação de letras
palavra_oculta = '*' * len(palavra_escolhida)

#Criação de lista vazia para armazenamento das tentativas
tentativa_letra = []
max_tentativas = 6

while True:
  #Exibe na tela a palavra oculta
  print(palavra_oculta)

  #Solicita ao jogador uma nova letra
  letra = input('Digite uma letra:  ')

  #Verificação de letra já digitada
  if letra in tentativa_letra:
    print('Você já digitou essa letra, tente outra vez!')
    continue

  #Adiciona a tentativa  na lsita de letras digitadas
  tentativa_letra.append(letra)

  #Verificação de tentativa com a palavra sorteada
  if letra in palavra_escolhida:
    lista = []
    for indice in range(len(palavra_escolhida)):
      if letra == palavra_escolhida[indice]:
        lista.append(letra)
      else:
        lista.append(palavra_oculta[indice])
    palavra_oculta = ''.join(lista)

  #quando a letra não está na palabra sorteada
  else:
    max_tentativas -= 1
    print(f'Letra não localizada! Você tem mais {max_tentativas} tentativas')

  #Verificação de vitória do jogo
  if palavra_oculta == palavra_escolhida:
    print('Parabéns! Você é o vencedor!')
    break
  elif max_tentativas == 0:
    print(f'VOCÊ PERDEU!!!')
    print(f'A palavra era {palavra_escolhida}.')
    break
```
